```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
x = pd.Series(range(-5, 50)) # or some other range of numbers for your x values

def squared_minus_x_plus_two(x):
    
    return x ** 2 - x + 2

y = x.apply(squared_minus_x_plus_two)
plt.annotate('Intersection', xy=(0, 0), xytext=(-3, 700),
             arrowprops={'facecolor': 'blue'})
plt.suptitle('y=x^2 - x + 2')
plt.plot(x, y)
```

```python
x = pd.Series(range(2, 50))
def sqroot_x(x):
    return math.sqrt(x)
y = x.apply(sqroot_x)
plt.suptitle('y=sqrt x')
plt.plot(x,y)


```

```python
x = pd.Series(range(2, 50))
def pow_x(x):
    return math.pow(x,3)
y=x.apply(pow_x)
plt.suptitle('y = x ^ 3')
plt.plot(x,y)
```

```python
x = pd.Series(range(2, 50))
def exp_x(x):
    return math.pow(2,x)
y=x.apply(exp_x)
plt.suptitle('2 ^ x')
plt.plot(x,y)
```
