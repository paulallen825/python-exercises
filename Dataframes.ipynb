```python
from pydataset import data
import pandas as pd
import numpy as np
```

```python
np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
print(df)
```

```python
#1.n Create a column named passing_english that indicates whether each student has a passing grade in english.
```

```python
df['passing_english'] = df.english >= 70
df
```

```python
#1.b Sort the english grades by the passing_english column. How are duplicates handled?
#It sorts by False and then True
```

```python
df.sort_values(by='passing_english')
```

```python
#1.c Sort the english grades first by passing_english and then by student name.
#All the students that are failing english should be first, 
#and within the students that are failing english they should be ordered alphabetically. 
#The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)
```

```python
df.sort_values(['passing_english','name'], ascending =(True, True))
```

```python
#1.d Sort the english grades first by passing_english, and then by the actual english grade, 
#similar to how we did in the last step.
```

```python
df.sort_values(['passing_english','english'], ascending =(False, False))
```

```python
#1.e Calculate each students overall grade and add it as a column on the dataframe. 
#The overall grade is the average of the math, english, and reading grades.
```

```python
df['overall_grade'] = (df.math+df.english+df.reading)/3
#=round((df.math + df.english + df.reading)/3,0)
#df[[‘English’, ‘reading’, ‘math’]].mean(axis=1)
df
```

```python
#b What are the data types of each column
mpg.dtypes
```

```python
#c Summarize the dataframe with .info and .describe
information = mpg.info
description = mpg.describe
print(information)
print(description)
```

```python
#d Rename the cty column to city.
#e Rename the hwy column to highway
mpg.rename(columns={'cty': 'city','hwy':'highway'})

#mpg.rename(columns={'cty': 'city','hwy':'highway', implace = True}) Mutates the df
```

```python
#f Do any cars have better city mileage than highway mileage? No
mpg['better_city'] = mpg.cty > mpg.hwy
mpg.sort_values('better_city',ascending = True)
```

```python
#g Create a column named mileage_difference this column 
#should contain the difference between highway and city mileage for each car.
mpg['mileage_difference'] = mpg.hwy-mpg.cty
mpg
```

```python
#h Which car (or cars) has the highest mileage difference?
mpg.sort_values('mileage_difference',ascending = False).head(5)
```

```python
#I Which compact class car has the lowest highway mileage? The best?
#lowest highway mileage - ford mustang
#best highway mileage - volkswagon new beetle
lowest_hw=mpg[mpg['class']=='subcompact'].sort_values(['hwy'], ascending=True).head(5)
best_hw=mpg[mpg['class']=='subcompact'].sort_values(['hwy'], ascending=False).head(5)
best_hw
```

```python
#J Create a column named average_mileage that is the mean of the city and highway mileage.
mpg['average_mileage'] = (mpg.cty+mpg.hwy)/2
mpg
```

```python
#k Which dodge car has the best average mileage? The worst?
#best average is index 38
best_average=mpg[mpg['manufacturer']=='dodge'].sort_values(['average_mileage'], ascending=False).head(1)
#best_average
worst_average=mpg[mpg['manufacturer']=='dodge'].sort_values(['average_mileage'], ascending=True).head(5)
worst_average
```

```python
#How many animals are hoppers that are above the median speed? What percentage is this?
```

```python
#found the median speed
median_speed =mammals['speed'].median()
median_speed
```

```python
#setting above median to true or false for animals above the median speed or not
above_median = mammals.speed >median_speed
above_median
```

```python
#counting the t/f values for above and below median speed
above_median.value_counts()
```

```python
#turning them into a percentage
above_median.value_counts(normalize=True).mul(100).round(2).astype(str)+'%'
```
